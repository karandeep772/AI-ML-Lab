```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
irisDF = pd.read_csv('IRIS.csv')
```

```python
irisDF.head(15)
```

```python
irisDF.describe()
```

```python
irisDF.info()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(irisDF.index, irisDF['sepal_length'],color='green', linestyle='-')
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.title("Sepal Length Line Plot")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.lineplot(data=irisDF, x=irisDF.index, y="sepal_length", hue="species")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.title("Sepal Length Line Plot")
plt.grid(True)
plt.show()
```

```python
irisDF.head(15).plot(kind="bar", y=["sepal_length", "sepal_width"], figsize=(12, 6), color=["blue", "green"])

# Labels and title
plt.xlabel("Index(first 15)")
plt.ylabel("Length (cm)")
plt.title("Comparison of Sepal Length and Sepal Width")
plt.legend(["Sepal Length", "Sepal Width"])
plt.show()
```

```python
sns.scatterplot(data=irisDF, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
sns.scatterplot(data=irisDF, x='petal_length', y='petal_width', hue='species')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
sns.countplot(data=irisDF, x='species', palette ="Set3")
plt.title('Count of Each Species')
plt.show()
```

```python
sns.boxplot(data=irisDF, x='species', y='sepal_length', color = "black")
plt.title('Box Plot of Sepal Length by Species')
plt.show()
```

```python
sns.boxplot(data=irisDF, x='species', y='petal_length', color = "grey")
plt.title('Box Plot of Petal Length by Species')
plt.show()
```

```python
numeric_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Plot histograms for each numeric column
for column in numeric_columns:
    plt.figure(figsize=(4,2))
    sns.histplot(data=irisDF, x=column, kde=True)
    plt.title(f'Histogram of {column}')
    plt.show()
```

```python
plt.figure(figsize=(6, 5))
corr = irisDF.corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt='.2f', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```
